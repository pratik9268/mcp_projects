```python
import pandas as pd
```

```python
df = pd.read_csv('recipes_data.csv')
```

```python
df.head()
```

```python
df.drop(columns=['source','site','ingredients'],inplace=True)
df.rename(columns={'NER':'Ingridents'},inplace=True)
df['Ingridents'][0]
# df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.mode().iloc[0]
```

```python
most_common_title = df['title'].mode()[0]
most_common_title
```

```python
df = df[df['title'] != most_common_title]
df
```

```python
df = df.sample(n=1000, random_state=42)  # Random 1000 rows
df
```

```python
cols = df.columns.tolist()
# Swap positions of 'link' and 'Ingridents'
link_index = cols.index('link')
ingr_index = cols.index('Ingridents')
cols[link_index], cols[ingr_index] = cols[ingr_index], cols[link_index]
df = df[cols]
```

```python
df.head()
```

```python
df.to_csv('recipe_cleaned.csv',index=False)
```
